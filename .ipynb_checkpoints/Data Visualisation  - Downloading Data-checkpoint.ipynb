### The CSV File Format
A simple way of storing data in a text file is to write the data as a series of values separated by commas, which is called comma-separated values. The resulting files is what is called CSV files.
CSV files - A comma-separated values files which allows data to be saved in a tabular format. 

Name,Email,Phone Number,Home Address,

Jane Doe,jane@example.com,923-456-7808,123 Church Street,

John Doe,john@example.com,578-765-6321,321 Mango Avenue

CSV files can be tricky for humans
to read, but they’re easy for programs to process and extract values from, which speeds up the data analysis process.

### Plotting Dates

```python
import csv
from datetime import datetime
import matplotlib.pyplot as plt

filename = 'data_files/the_csv_file_format/data/sitka_weather_2018_simple.csv'

with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    print(header_row)

    for index,column_header in enumerate(header_row):
        print(index,column_header)
        
    ## get high temparatures for each day
    dates, highs = [], []
    
    for row in reader:
        current_date = datetime.strptime(row[2],'%Y-%m-%d')
        high = int(row[5])
        dates.append(current_date)
        highs.append([high])
#     print(highs)
    
## plots high temparatures
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.plot(dates,highs,c='red')

# Format plot.
plt.title('Daily temparatures. July, 2018', fontsize=24)
plt.xlabel('',fontsize=16)
fig.autofmt_xdate()
plt.ylabel('', fontsize=16)
plt.tick_params(axis="both",which='major',labelsize=16)

plt.show()
```

### Plotting a Second Data Series

We can make our informative graph even more useful by including the low temperatures. We need to extract the low temperatures from the data file and then add them to our graph

```python
import csv
from datetime import datetime

from matplotlib import pyplot as plt

basepath = 'data_files/'
filename = basepath +'the_csv_file_format/data/sitka_weather_2018_simple.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)

    # Get dates, and high and low temperatures from this file
    dates, highs,lows = [], [], []
    for row in reader:
        current_date = datetime.strptime(row[2], '%Y-%m-%d')
        dates.append(current_date)
        high = int(row[5])
        low = int(row[6])
        highs.append(high)
        lows.append(low)

# Plot the high temperatures.
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.plot(dates, highs, c='red')
ax.plot(dates, lows, c='blue')

# Format plot.
plt.title("Daily high and low temperatures - 2018", fontsize=24)
plt.xlabel('Dates', fontsize=16)
fig.autofmt_xdate()
plt.ylabel("Temperature (F)", fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=16)

plt.show()
```

### Shading an Area in the Chart
Now that we've added two data series, we can now examine the range of temperatures for each day. Let’s add a finishing touch to the graph by using shading to show the range between each day’s high and low temperatures. To do so, we’ll use the fill_between() method, which takes a series of x-values and two series of y-values, and fills the space between the two y-value series

```python
import csv
from datetime import datetime

from matplotlib import pyplot as plt

basepath = 'data_files/'
filename = basepath +'the_csv_file_format/data/sitka_weather_2018_simple.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)

    # Get dates, and high and low temperatures from this file
    dates, highs,lows = [], [], []
    for row in reader:
        current_date = datetime.strptime(row[2], '%Y-%m-%d')
        dates.append(current_date)
        high = int(row[5])
        low = int(row[6])
        highs.append(high)
        lows.append(low)

# Plot the high temperatures.
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.plot(dates, highs, c='red', alpha=.5)
ax.plot(dates, lows, c='blue', alpha=0.5)
plt.fill_between(dates,highs,lows,facecolor='blue',alpha=0.1)


# Format plot.
plt.title("Daily high and low temperatures - 2018", fontsize=24)
plt.xlabel('Dates', fontsize=16)
fig.autofmt_xdate()
plt.ylabel("Temperature (F)", fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=16)

plt.show()
```

```python
import csv

basepath = 'data_files/'
filename = basepath +'the_csv_file_format/data/death_valley_2018_simple.csv' 
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    # Get dates, and high and low temperatures from this file. 
    dates, highs, lows = [], [], []
    for row in reader:
        
        try:  
            current_date = datetime.strptime(row[2], '%Y-%m-%d')
            high = int(row[4])
            low = int(row[5])
        except ValueError:
            print(f"Missing data for {current_date}")
        else:
            dates.append(current_date)
            highs.append(high) 
            lows.append(low)

        # Plot the high temperatures.
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.plot(dates, highs, c='red', alpha=0.5)
ax.plot(dates, lows, c='blue', alpha=0.5)
plt.fill_between(dates,highs,lows,facecolor='blue',alpha=0.1)

title = "Daily high and low temperatures - 2018\nDeath Valley, CA" 
plt.title(title, fontsize=20)
plt.xlabel('', fontsize=16)
fig.autofmt_xdate()

plt.ylabel("Temperature (F)", fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=16)

plt.show()
```

### Extracting Location Data
Next, we’ll pull the location data for each earthquake, and then we can make a map of the earthquakes.
The location data is stored under the key "geometry". Inside the geometry dictionary is a "coordinates" key, and the first two values in this list are the longitude and latitude.

```python
import json

# Explore the structure of the data. 
basepath = 'data_files/'
filename =  basepath +'mapping_global_data_sets/data/eq_data_1_day_m1.json' 
with open(filename) as f:
    all_eq_data = json.load(f)
readable_file = basepath +'mapping_global_data_sets/data/readable_eq_data.json' 
with open(readable_file, 'w') as f:
    json.dump(all_eq_data, f, indent=4)

all_eq_dicts = all_eq_data['features'] 
mags, longs, lats = [], [], []
for eq_dict in all_eq_dicts:
    mag = eq_dict['properties']['mag']
    long  = eq_dict['geometry']['coordinates'][0]
    lat  = eq_dict['geometry']['coordinates'][1]
    mags.append(mag)
    longs.append(long)
    lats.append(lat)
print(mags[:10]) 
print(longs[:5])
print(lats[:5])

```

### Building a World Map
With the information we’ve pulled so far, we can build a simple world map. Although it won’t look presentable yet, we want to make sure the information is displayed correctly before focusing on style and presentation issues.

```python
import json
from plotly.graph_objs import Scattergeo, Layout
from plotly import offline

with open(filename) as f:
    all_eq_data = json.load(f)
readable_file = basepath +'mapping_global_data_sets/data/readable_eq_data.json' 
with open(readable_file, 'w') as f:
    json.dump(all_eq_data, f, indent=4)
              
all_eq_dicts = all_eq_data['features']             
# all_eq_dicts = all_eq_data['features'] 
mags, longs, lats = [], [], []
for eq_dict in all_eq_dicts:
    mag = eq_dict['properties']['mag']
    long  = eq_dict['geometry']['coordinates'][0]
    lat  = eq_dict['geometry']['coordinates'][1]
    mags.append(mag)
    longs.append(long)
    lats.append(lat)
data = [Scattergeo(lon=longs,lat=lats)]
my_layout = Layout(title='Global Earthquakes')

fig = {'data':data, 'layout':my_layout}

offline.plot(fig, filename='quakes.html')
```

#### Customizing Marker Size

The current map shows the location of each earthquake, but it doesn’t communicate the severity of any earthquake. We want viewers to immediately see where the most significant earthquakes occur in the world

```python
import json
from plotly.graph_objs import Scattergeo, Layout
from plotly import offline

basepath = 'data_files/'
filename =  basepath +'mapping_global_data_sets/data/eq_data_1_day_m1.json' 
with open(filename) as f:
    all_eq_data = json.load(f)
readable_file = basepath +'mapping_global_data_sets/data/readable_eq_data.json' 
with open(readable_file, 'w') as f:
    json.dump(all_eq_data, f, indent=4)
              
all_eq_dicts = all_eq_data['features']             
# all_eq_dicts = all_eq_data['features'] 
mags, longs, lats = [], [], []
for eq_dict in all_eq_dicts:
    mag = eq_dict['properties']['mag']
    long  = eq_dict['geometry']['coordinates'][0]
    lat  = eq_dict['geometry']['coordinates'][1]
    mags.append(mag)
    longs.append(long)
    lats.append(lat)
    
#### A Different Way of Specifying Chart Data
data = [{
    'type': 'scattergeo',
    'lon': longs,
    'lat' : lats,
    'marker' : {
        'size': [5*mag for mag in mags],
    },
}]


my_layout = Layout(title='Global Earthquakes')

fig = {'data':data, 'layout':my_layout}

offline.plot(fig, filename='quake.html')
```

#### Customizing Marker Colors
Now, let's try to customize each marker’s color to provide some classification to the severity of each earthquake. We’ll use Plotly’s colorscales to do this. We''' make use of the  eq_data_30_day_m1.json file. This file includes earthquake data for a 30-day period, and the map will be much more interesting to look at using this larger data set. 

```python
import json
from plotly.graph_objs import Scattergeo, Layout
from plotly import offline

basepath = 'data_files/'
filename =  basepath +'mapping_global_data_sets/data/eq_data_30_day_m1.json' 
with open(filename) as f:
    all_eq_data = json.load(f)
readable_file = basepath +'mapping_global_data_sets/data/readable_eq_data.json' 
with open(readable_file, 'w') as f:
    json.dump(all_eq_data, f, indent=4)
              
all_eq_dicts = all_eq_data['features']             
# all_eq_dicts = all_eq_data['features'] 
mags, longs, lats = [], [], []
for eq_dict in all_eq_dicts:
    mag = eq_dict['properties']['mag']
    long  = eq_dict['geometry']['coordinates'][0]
    lat  = eq_dict['geometry']['coordinates'][1]
    mags.append(mag)
    longs.append(long)
    lats.append(lat)
    
#### A Different Way of Specifying Chart Data
data = [{
    'type': 'scattergeo',
    'lon': longs,
    'lat' : lats,
    'marker' : {
        'size': [5*mag for mag in mags],
    'color':mags,
    'colorscale': 'Viridis',
    'reversescale': True,
    'colorbar':{'title':'Magnitude'},
    },
}]


my_layout = Layout(title='Global Earthquakes')

fig = {'data':data, 'layout':my_layout}

offline.plot(fig, filename='quake.html')
```

### Adding Hover Text
To finish this map, we’ll add some informative text that appears when you hover over the marker representing an earthquake. In addition to showing the longitude and latitude, which appear by default, we’ll show the magnitude and provide a description of the approximate location as well.

```python
import json
from plotly.graph_objs import Scattergeo, Layout
from plotly import offline

basepath = 'data_files/'
filename =  basepath +'mapping_global_data_sets/data/eq_data_30_day_m1.json' 
with open(filename) as f:
    all_eq_data = json.load(f)
readable_file = basepath +'mapping_global_data_sets/data/readable_eq_data.json' 
with open(readable_file, 'w') as f:
    json.dump(all_eq_data, f, indent=4)
              
all_eq_dicts = all_eq_data['features']             
# all_eq_dicts = all_eq_data['features'] 
mags, longs, lats, hover_texts = [], [], [],[]
for eq_dict in all_eq_dicts:
    mag = eq_dict['properties']['mag']
    long  = eq_dict['geometry']['coordinates'][0]
    lat  = eq_dict['geometry']['coordinates'][1]
    title = eq_dict['properties']['title']
    mags.append(mag)
    longs.append(long)
    lats.append(lat)
    hover_texts.append(title)
    
#### A Different Way of Specifying Chart Data
data = [{
    'type': 'scattergeo',
    'lon': longs,
    'lat' : lats,
    'text': hover_texts,
    'marker' : {
        'size': [5*mag for mag in mags],
    'color':mags,
    'colorscale': 'Viridis',
    'reversescale': True,
    'colorbar':{'title':'Magnitude'},
    },
}]


my_layout = Layout(title='Global Earthquakes')

fig = {'data':data, 'layout':my_layout}

offline.plot(fig, filename='quake.html')
```

```python
import json
from plotly.graph_objs import Scattergeo, Layout
from plotly import offline

basepath = 'data_files/'
filename =  basepath +'mapping_global_data_sets/data/eq_data_30_day_m1.json' 
with open(filename) as f:
    all_eq_data = json.load(f)
readable_file = basepath +'mapping_global_data_sets/data/readable_eq_data.json' 
with open(readable_file, 'w') as f:
    json.dump(all_eq_data, f, indent=4)
              
all_eq_dicts = all_eq_data['features']             
# all_eq_dicts = all_eq_data['features'] 
mags, longs, lats, hover_texts = [], [], [],[]
for eq_dict in all_eq_dicts:
#     refactor here
    mags.append(eq_dict['properties']['mag'])
    longs.append(eq_dict['geometry']['coordinates'][0])
    lats.append(eq_dict['geometry']['coordinates'][1])
    hover_texts.append(eq_dict['properties']['title'])
    
#### A Different Way of Specifying Chart Data
data = [{
    'type': 'scattergeo',
    'lon': longs,
    'lat' : lats,
    'text': hover_texts,
    'marker' : {
        'size': [5*mag for mag in mags],
    'color':mags,
    'colorscale': 'Viridis',
    'reversescale': True,
    'colorbar':{'title':'Magnitude'},
    },
}]

# Automated title
my_layout = Layout(title=all_eq_data['metadata']['title'])

fig = {'data':data, 'layout':my_layout}

offline.plot(fig, filename='quake.html')
```
